```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set(style='whitegrid')
plt.rcParams["figure.figsize"] = (12, 6)

# Load datasets
business_df = pd.read_csv('data_filtered/business_filtered.csv')
user_df = pd.read_csv('data_filtered/user_small.csv')
checkin_df = pd.read_csv('data_filtered/checkin_small.csv')
review_df = pd.read_csv('data_filtered/review_small_escaped.csv')
tips_df = pd.read_csv('data_filtered/tips_small.csv')
```

```python
# =================== BUSINESS ===================
print("Business Dataset Overview:")
print(business_df.info())
print(business_df.describe())
print(business_df.isnull().sum())

# Top 10 cities with most businesses
top_cities = business_df['city'].value_counts().head(10)
sns.barplot(x=top_cities.values, y=top_cities.index)
plt.title("Top 10 Cities with Most Businesses")
plt.xlabel("Number of Businesses")
plt.ylabel("City")
plt.show()

# Stars distribution
sns.histplot(business_df['stars'], bins=20, kde=True)
plt.title("Business Star Ratings Distribution")
plt.xlabel("Stars")
plt.ylabel("Frequency")
plt.show()
```

```python
# =================== USER ===================
print("\nUser Dataset Overview:")
print(user_df.info())
print(user_df.describe())
print(user_df.isnull().sum())

# Review count distribution
sns.histplot(user_df['review_count'], bins=50, log_scale=True)
plt.title("Distribution of User Review Counts")
plt.xlabel("Review Count")
plt.ylabel("Number of Users")
plt.show()

# Average stars distribution
sns.histplot(user_df['average_stars'], bins=30, kde=True)
plt.title("Distribution of Users' Average Stars")
plt.xlabel("Average Stars")
plt.ylabel("Frequency")
plt.show()
```

```python
# =================== CHECK-IN ===================
print("\nCheck-in Dataset Overview:")
print(checkin_df.info())
print(checkin_df.head())

# Number of check-ins per business
checkin_counts = checkin_df['business_id'].value_counts()
sns.histplot(checkin_counts, bins=30, log_scale=True)
plt.title("Number of Check-ins per Business")
plt.xlabel("Check-in Count")
plt.ylabel("Number of Businesses")
plt.show()
```

```python
# =================== REVIEW ===================
print("\nReview Dataset Overview:")
print(review_df.info())
print(review_df.describe())
print(review_df.isnull().sum())

# Review star distribution
sns.countplot(x='stars', data=review_df)
plt.title("Distribution of Review Star Ratings")
plt.xlabel("Stars")
plt.ylabel("Count")
plt.show()

# Review length distribution
review_df['review_length'] = review_df['text'].str.len()
sns.histplot(review_df['review_length'], bins=50)
plt.title("Distribution of Review Text Lengths")
plt.xlabel("Length of Review")
plt.ylabel("Frequency")
plt.show()
```

```python
# =================== TIPS ===================
print("\nTips Dataset Overview:")
print(tips_df.info())
print(tips_df.describe())
print(tips_df.isnull().sum())

# Tip length distribution
tips_df['tip_length'] = tips_df['text'].str.len()
sns.histplot(tips_df['tip_length'], bins=50)
plt.title("Distribution of Tip Text Lengths")
plt.xlabel("Tip Length")
plt.ylabel("Frequency")
plt.show()
```
